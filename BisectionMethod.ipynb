```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 

def f(x):
    return (x**3-4*x-9)

def bisection_method(f,x1,x2):
    tol=1e-5
    max_iter=50

    prev_mid=None
    iterations=[]

    for i in range(max_iter):
        mid=(x1+x2)/2
        f_mid=f(mid)

        if prev_mid!=None:
            abs_error=(abs(mid-prev_mid)/mid)*100
        else:
            abs_error=None
        iterations.append([i+1,x1,x2,mid,f_mid,(x2-x1),abs_error])

        if abs(f_mid)<tol or (abs(x2-x1)/2)<tol:
            break


        if f(x1)*f_mid<0:
            x2=mid
        else:
            x1=mid

        prev_mid=mid
    
    return iterations

x1=2
x2=3

results=bisection_method(f,x1,x2)

columns=["Iterations","x1","x2","mid","f(mid)","Interval","Absolute Error"]
df=pd.DataFrame(results,columns=columns)

print(df)

x_vals=np.linspace(x1-1,x2+1,50)
y_vals=f(x_vals)

plt.figure(figsize=(10,6))
plt.plot(x_vals,y_vals,color="blue",label=f"f(x)=x^3-4x-9")
plt.axhline(0,color="black",linestyle="--",label=f"y={0}")

for index,row in df.iterrows():
    x_start=row["x1"]
    x_end=row["x2"]
    y_start=f(x_start)
    y_end=f(x_end)

    plt.plot([x_start,x_end],[y_start,y_end],"ro--",label=f"First Interval" if index==0 else "")

plt.title("Bisection With Graph")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()
plt.show()
```
